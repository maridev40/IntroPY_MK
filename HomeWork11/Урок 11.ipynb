f(x) = -5*x**2 + 10*x - 150
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
import math
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
fx = -5*x**2 + 10*x - 150
fx
```

```python
p = plot(fx, (x, -10, 10))
```

```python
sqrt1, sqrt2 = solve(fx) # корни
print(sqrt1, sqrt2)
```

```python
apex = solve(diff(fx))[0] # вершина
print(apex)
```

```python
solve(diff(fx) > 0) # интервалы, на которых функция возврастает
```

```python
solve(diff(fx) < 0) # интервалы, на которых функция убывает
```

```python
solve(fx > 0) # промежутки, на котором f > 0
```

```python
solve(fx < 0) # промежутки, на котором f < 0
```
